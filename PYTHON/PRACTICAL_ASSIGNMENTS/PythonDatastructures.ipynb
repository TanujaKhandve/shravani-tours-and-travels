```python
# Linear Search
def linear_search(a, x):
    for i in range(len(a)):
        if a[i] == x:
            return i
    return -1

# Binary Search (sorted list required)
def binary_search(a, x):
    low, high = 0, len(a)-1
    while low <= high:
        mid = (low+high)//2
        if a[mid] == x:
            return mid
        elif a[mid] < x:
            low = mid+1
        else:
            high = mid-1
    return -1

nums = [5, 3, 7, 1, 9]
print("Linear Search:", linear_search(nums, 7))
print("Binary Search:", binary_search(sorted(nums), 7))
```

```python
def bubble_sort(a):
    for i in range(len(a)):
        for j in range(len(a)-i-1):
            if a[j] > a[j+1]: a[j], a[j+1] = a[j+1], a[j]
    return a

def insertion_sort(a):
    for i in range(1, len(a)):
        k, j = a[i], i-1
        while j >= 0 and a[j] > k:
            a[j+1] = a[j]; j -= 1
        a[j+1] = k
    return a

def quick_sort(a):
    return a if len(a) <= 1 else quick_sort([x for x in a[1:] if x <= a[0]]) + [a[0]] + quick_sort([x for x in a[1:] if x > a[0]])

def merge_sort(a):
    if len(a) <= 1: return a
    l, r = merge_sort(a[:len(a)//2]), merge_sort(a[len(a)//2:])
    res=[]
    while l and r: res.append((l if l[0] < r[0] else r).pop(0))
    return res+l+r

nums = [5, 2, 9, 1, 7]
print("Bubble:", bubble_sort(nums[:]))
print("Insertion:", insertion_sort(nums[:]))
print("Quick:", quick_sort(nums[:]))
print("Merge:", merge_sort(nums[:]))
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def create(self, data_list):
        for data in data_list:
            self.insert_end(data)

    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def delete(self, key):
        temp = self.head
        if temp and temp.data == key:
            self.head = temp.next
            return
        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next
        if temp:
            prev.next = temp.next

    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                return True
            temp = temp.next
        return False

    def reverse(self):
        prev = None
        current = self.head
        while current:
            nxt = current.next
            current.next = prev
            prev = current
            current = nxt
        self.head = prev

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# Example usage
ll = SinglyLinkedList()
ll.create([10, 20, 30])
print("Original list:")
ll.display()
ll.insert_end(40)
print("After inserting 40:")
ll.display()
ll.delete(20)
print("After deleting 20:")
ll.display()
print("Search 30:", ll.search(30))
ll.reverse()
print("After reversing:")
ll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyCircularLinkedList:
    def __init__(self):
        self.head = None

    def create(self, data_list):
        for data in data_list:
            self.insert_end(data)

    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
            self.head.prev = self.head
            return
        tail = self.head.prev
        tail.next = new_node
        new_node.prev = tail
        new_node.next = self.head
        self.head.prev = new_node

    def delete(self, key):
        if not self.head:
            return
        curr = self.head
        while True:
            if curr.data == key:
                if curr.next == curr:  # only one node
                    self.head = None
                    return
                curr.prev.next = curr.next
                curr.next.prev = curr.prev
                if curr == self.head:
                    self.head = curr.next
                return
            curr = curr.next
            if curr == self.head:
                break

    def display(self):
        if not self.head:
            print("List is empty")
            return
        curr = self.head
        while True:
            print(curr.data, end=" ")
            curr = curr.next
            if curr == self.head:
                break
        print()

    def search(self, key):
        if not self.head:
            return False
        curr = self.head
        while True:
            if curr.data == key:
                return True
            curr = curr.next
            if curr == self.head:
                break
        return False

# Example
dll = DoublyCircularLinkedList()
dll.create([10, 20, 30])
dll.display()
dll.insert_end(40)
dll.display()
dll.delete(20)
dll.display()
print("Search 30:", dll.search(30))
print("Search 50:", dll.search(50))
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_sorted_lists(l1, l2):
    dummy = Node(0)
    tail = dummy
    while l1 and l2:
        if l1.data < l2.data:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next
    tail.next = l1 or l2
    return dummy.next

def display(head):
    while head:
        print(head.data, end=" -> ")
        head = head.next
    print("None")

# Example
a = Node(1); a.next = Node(3); a.next.next = Node(5)
b = Node(2); b.next = Node(4); b.next.next = Node(6)

print("List 1:")
display(a)
print("List 2:")
display(b)

merged = merge_sorted_lists(a, b)
print("Merged List:")
display(merged)
```

```python
# Assignment 6: Static & Dynamic Stack

# ---------- Static Stack ----------
class StaticStack:
    def __init__(self, size):
        self.stack = []
        self.size = size

    def push(self, item):
        if self.is_full():
            print("Static Stack is FULL")
        else:
            self.stack.append(item)

    def pop(self):
        if self.is_empty():
            print("Static Stack is EMPTY")
        else:
            return self.stack.pop()

    def is_empty(self):
        return len(self.stack) == 0

    def is_full(self):
        return len(self.stack) == self.size


# ---------- Dynamic Stack ----------
class DynamicStack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            print("Dynamic Stack is EMPTY")
        else:
            return self.stack.pop()

    def is_empty(self):
        return len(self.stack) == 0


# ---------- Demo ----------
# Static Stack
s1 = StaticStack(3)
s1.push(10)
s1.push(20)
s1.push(30)
s1.push(40)  # Will show FULL
print("Static Pop:", s1.pop())

# Dynamic Stack
s2 = DynamicStack()
s2.push(100)
s2.push(200)
s2.push(300)
print("Dynamic Pop:", s2.pop())
```

```python
# Assignment 7: Applications of Stack

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0


# 1. Reverse string and check palindrome
def reverse_string(s):
    st = Stack()
    for ch in s:
        st.push(ch)
    rev = ""
    while not st.is_empty():
        rev += st.pop()
    return rev

def is_palindrome(s):
    return s == reverse_string(s)


# 2. Infix to Postfix conversion
def infix_to_postfix(expression):
    precedence = {'+':1, '-':1, '*':2, '/':2, '^':3}
    st = Stack()
    output = ""
    for ch in expression:
        if ch.isalnum():
            output += ch
        elif ch == '(':
            st.push(ch)
        elif ch == ')':
            while not st.is_empty() and st.peek() != '(':
                output += st.pop()
            st.pop()
        else:  # operator
            while (not st.is_empty()) and st.peek() != '(' and precedence[ch] <= precedence[st.peek()]:
                output += st.pop()
            st.push(ch)
    while not st.is_empty():
        output += st.pop()
    return output


# 3. Postfix Evaluation
def eval_postfix(expression):
    st = Stack()
    for ch in expression:
        if ch.isdigit():
            st.push(int(ch))
        else:
            b = st.pop()
            a = st.pop()
            if ch == '+': st.push(a + b)
            elif ch == '-': st.push(a - b)
            elif ch == '*': st.push(a * b)
            elif ch == '/': st.push(a // b)
    return st.pop()


# ---- MAIN ----
# Reverse & Palindrome
s = "madam"
rev = reverse_string(s)
print("Original:", s)
print("Reversed:", rev)
print("Palindrome?", is_palindrome(s))

# Infix to Postfix + Evaluation
infix_exp = "3+(2*4)"
postfix_exp = infix_to_postfix(infix_exp)
print("\nInfix:", infix_exp)
print("Postfix:", postfix_exp)
print("Postfix Evaluation:", eval_postfix(postfix_exp))
```

```python
# Assignment 8: Linear Queue - Static & Dynamic Implementation
##########SLIP NO 1 Q2 STATIC implementation
# ----- Static Queue Implementation -----
class StaticQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = 0
        self.rear = -1

    def isEmpty(self):
        return self.front > self.rear

    def isFull(self):
        return self.rear == self.size - 1

    def enqueue(self, item):
        if self.isFull():
            print("Static Queue is Full!")
        else:
            self.rear += 1
            self.queue[self.rear] = item
            print(f"Enqueued: {item}")

    def dequeue(self):
        if self.isEmpty():
            print("Static Queue is Empty!")
        else:
            item = self.queue[self.front]
            self.front += 1
            print(f"Dequeued: {item}")

# ----- Dynamic Queue Implementation -----
class DynamicQueue:
    def __init__(self):
        self.queue = []

    def isEmpty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)
        print(f"Enqueued: {item}")

    def dequeue(self):
        if self.isEmpty():
            print("Dynamic Queue is Empty!")
        else:
            item = self.queue.pop(0)
            print(f"Dequeued: {item}")

# ----- Main Program -----
print("----- Static Queue -----")
sq = StaticQueue(3)
sq.enqueue(10)
sq.enqueue(20)
sq.enqueue(30)
sq.enqueue(40)  # Overflow
sq.dequeue()
sq.dequeue()
sq.dequeue()
sq.dequeue()    # Underflow

print("\n----- Dynamic Queue -----")
dq = DynamicQueue()
dq.enqueue(100)
dq.enqueue(200)
dq.enqueue(300)
dq.dequeue()
dq.dequeue()
dq.dequeue()
dq.dequeue()    # Underflow
```

```python
# ----- Circular Queue Implementation -----
class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = self.rear = -1  ## IT means queue is empty

    def isEmpty(self):
        return self.front == -1

    def isFull(self):
        return (self.rear + 1) % self.size == self.front ## IT means queue is full How to check if queue is full that is rear+1 is equal to front

    def enqueue(self, item):
        if self.isFull():
            print("Circular Queue is Full!")
        else:
            if self.front == -1:
                self.front = 0
            self.rear = (self.rear + 1) % self.size
            self.queue[self.rear] = item
            print(f"Enqueued: {item}")

    def dequeue(self):
        if self.isEmpty():
            print("Circular Queue is Empty!")
        else:
            item = self.queue[self.front]
            if self.front == self.rear:
                self.front = self.rear = -1
            else:
                self.front = (self.front + 1) % self.size
            print(f"Dequeued: {item}")


# ----- Priority Queue Implementation -----
class PriorityQueue:
    def __init__(self):
        self.queue = []

    def isEmpty(self):
        return len(self.queue) == 0

    def enqueue(self, item, priority):
        self.queue.append((priority, item))
        self.queue.sort(reverse=True)  # Higher priority first
        print(f"Enqueued: {item} with priority {priority}")

    def dequeue(self):
        if self.isEmpty():
            print("Priority Queue is Empty!")
        else:
            priority, item = self.queue.pop(0)
            print(f"Dequeued: {item} with priority {priority}")


# ----- Main Program -----
print("----- Circular Queue -----")
cq = CircularQueue(3)
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)
cq.enqueue(4)  # Overflow
cq.dequeue()
cq.enqueue(4)
cq.dequeue()
cq.dequeue()
cq.dequeue()
cq.dequeue()  # Underflow

print("\n----- Priority Queue -----")
pq = PriorityQueue()
pq.enqueue("Apple", 2)
pq.enqueue("Banana", 1)
pq.enqueue("Cherry", 3)
pq.dequeue()
pq.dequeue()
pq.dequeue()
pq.dequeue()  # Underflow
```

```python
# Assignment 10: Tree Traversals and Operations

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

# ----- Insertion in Binary Tree -----
def insert(root, key):
    if root is None:
        return Node(key)
    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)
    return root

# ----- Inorder Traversal (LNR) -----
def inorder(root):
    if root:
        inorder(root.left)
        print(root.key, end=" ")
        inorder(root.right)

# ----- Preorder Traversal (NLR) -----
def preorder(root):
    if root:
        print(root.key, end=" ")
        preorder(root.left)
        preorder(root.right)

# ----- Postorder Traversal (LRN) -----
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.key, end=" ")

# ----- Search in Binary Tree -----
def search(root, key):
    if root is None:
        return False
    if root.key == key:
        return True
    elif key < root.key:
        return search(root.left, key)
    else:
        return search(root.right, key)

# ----- Main Program -----
root = None
values = [50, 30, 20, 40, 70, 60, 80]

for v in values:
    root = insert(root, v)

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)

print("\n\nSearching 40:", "Found" if search(root, 40) else "Not Found")
print("Searching 100:", "Found" if search(root, 100) else "Not Found")
```

```python
# Assignment 11: Calculate indegree and outdegree of a given graph

# Function to calculate indegree and outdegree
def calculate_degrees(graph):
    vertices = list(graph.keys())
    indegree = {v: 0 for v in vertices}
    outdegree = {v: len(graph[v]) for v in vertices}

    # Calculate indegree
    for v in vertices:
        for neighbor in graph[v]:
            indegree[neighbor] += 1

    return indegree, outdegree


# ----- Main Program -----
# Graph representation as adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['C'],
    'C': ['A'],
    'D': ['C']
}

print("Graph:", graph)

indegree, outdegree = calculate_degrees(graph)

print("\nVertex\tIndegree\tOutdegree")
for vertex in graph:
    print(f"{vertex}\t{indegree[vertex]}\t\t{outdegree[vertex]}")
```
