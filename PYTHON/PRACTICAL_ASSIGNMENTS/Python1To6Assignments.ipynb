Q2: Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the values entered are duplicates otherwise it prints "ALL UNIQUE".
Code Snippet:-


```python
values=[]

for i in range(6):
    num=int(input(f"Enter integer {i+1}: "))
    values.append(num)
    
if len(set(values))<len(values):
    print("DUPLICATES")
else:
    print("ALL UNIQUE")

```

Q2: Write a python function to calculate the factorial of a number. The function accept the number as an argument.

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

# Example
num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

Q3: Write a python function to check whether a number falls within a given range.

```python
def in_range(num, start, end):
    return start <= num <= end

# Example
n = int(input("Enter number: "))
s = int(input("Enter range start: "))
e = int(input("Enter range end: "))
print("In range?", in_range(n, s, e))
```

```python
def unique_list(lst):
    return list(set(lst))

# Example
sample = [1, 2, 2, 3, 3, 3, 3, 4, 5]
print("Sample list:", sample)
print("Unique list:", unique_list(sample))
```
